```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('data.csv')
filmler = veriSeti['Title']
turler = veriSeti['Genres']
rating = veriSeti['Average Rating']
oySayisi = veriSeti['Num Votes']
yil = veriSeti['Release Year']

# Türlerine göre rating sayisi
plt.figure(figsize=(30,10))  # Grafiğin boyutunu ayarlayın
plt.bar(filmler, turler)
plt.xlabel('Filmler')
plt.ylabel('Türler')
plt.title('Türlerine Göre Filmler')
plt.xticks(rotation=90)  # Film isimlerinin dikey yazılması
plt.show()
```

```python
ratingOrani = veriSeti['Average Rating'] / veriSeti['Num Votes']

plt.figure(figsize=(30,10))  # Grafiğin boyutunu ayarlayın
plt.bar(filmler, ratingOrani)
plt.xlabel('Filmler')
plt.ylabel('Rating Sayısının Oy Sayısına Oranı')
plt.title('Filmlerin Rating Sayısının Oy Sayısına Oranı')
plt.xticks(rotation=90)  # Film isimlerinin dikey yazılması
plt.show()


```
