```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('data.csv')
filmler = veriSeti['Title']
turler = veriSeti['Genres']
ortDerecelendirme = veriSeti['Average Rating']
oySayisi = veriSeti['Num Votes']
yil = veriSeti['Release Year']

plt.figure(figsize=(20, 7))  # Grafiin boyutunu ayarlayn
plt.bar(filmler, oySayisi)
plt.xlabel('Filmler')
plt.ylabel('Oylar')
plt.title('Oy Saysna Gre En iyi Filmler')
plt.xticks(rotation=90)  # Film isimlerinin dikey yazlmas
plt.show()
```

```python
ortDerecelendirme = veriSeti['Num Votes'] / veriSeti['Average Rating']

plt.figure(figsize=(20, 7))  # Grafiin boyutunu ayarlayn
plt.bar(filmler, ortDerecelendirme)
plt.xlabel('Filmler')
plt.ylabel('Oy Saysnn Ortalama Derecelendirmeye Oran')
plt.title('Filmlerin Oy Saysnn Ortalama Derecelendirmeye Oran')
plt.xticks(rotation=90)  # Film isimlerinin dikey yazlmas
plt.show()
```

```python
ratingOrani = veriSeti['Average Rating'] / veriSeti['Num Votes']

plt.figure(figsize=(20, 7))  # Grafiin boyutunu ayarlayn
plt.bar(filmler, ratingOrani)
plt.xlabel('Filmler')
plt.ylabel('Rating Saysnn Oy Saysna Oran')
plt.title('Filmlerin Rating Saysnn Oy Saysna Oran')
plt.xticks(rotation=90)  # lke isimlerinin dikey yazlmas
plt.show()
```
