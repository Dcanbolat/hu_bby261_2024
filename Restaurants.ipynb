```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('WorldsBestRestaurants.csv')
yil = veriSeti['Year']
siralama = veriSeti['Rank']
restaurantlar = veriSeti['Restaurant']
konum = veriSeti['Location']
ülke = veriSeti['Country']

plt.figure(figsize=(20, 7))  # Grafiğin boyutunu ayarlayın
plt.bar(restaurantlar, siralama)
plt.xlabel('Restaurantlar')
plt.ylabel('Sıralama')
plt.title('Restaurantların Sıralaması')
plt.xticks(rotation=90)  # Restaurant isimlerinin dikey yazılması
plt.show()
```

```python
yilSiralamaORani = veriSeti['Year'] / veriSeti['Rank']

plt.figure(figsize=(20, 7))  # Grafiğin boyutunu ayarlayın
plt.bar(restaurantlar, yilSiralamaORani)
plt.xlabel('Restaurantlar')
plt.ylabel('Yıllara Göre Sıralama')
plt.title('Restaurantların Yıllara Göre Sıralanma Oranı')
plt.xticks(rotation=90)  # Film isimlerinin dikey yazılması
plt.show()
```
