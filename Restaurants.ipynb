```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('WorldsBestRestaurants.csv')
yil = veriSeti['Year']
siralama = veriSeti['Rank']
restaurantlar = veriSeti['Restaurant']
konum = veriSeti['Location']
lke = veriSeti['Country']

plt.figure(figsize=(20, 7))  # Grafiin boyutunu ayarlayn
plt.bar(restaurantlar, siralama)
plt.xlabel('Restaurantlar')
plt.ylabel('Sralama')
plt.title('Restaurantlarn Sralamas')
plt.xticks(rotation=90)  # Restaurant isimlerinin dikey yazlmas
plt.show()
```

```python
yilSiralamaORani = veriSeti['Year'] / veriSeti['Rank']

plt.figure(figsize=(20, 7))  # Grafiin boyutunu ayarlayn
plt.bar(restaurantlar, yilSiralamaORani)
plt.xlabel('Restaurantlar')
plt.ylabel('Yllara Gre Sralama')
plt.title('Restaurantlarn Yllara Gre Sralanma Oran')
plt.xticks(rotation=90)  # Film isimlerinin dikey yazlmas
plt.show()
```
